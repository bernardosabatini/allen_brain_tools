# Initialize and find the ImageDataSet ids for the genes of interest

```python
import requests
import json

genes=['Adora2a', 'Tph2', 'Slc17a7']
section_orientation='coronal'

imageDataSet_ids={}

for gene in genes:
    build_string=''.join(['http://api.brain-map.org/api/v2/data/',
                      'query.json?criteria=model::SectionDataSet,',
                      'rma::criteria,',
                      "[failed$eq'false'],",
                      "products[abbreviation$eq'Mouse'],",
                      "plane_of_section[name$eq'",
                      section_orientation,
                      "'],",
                      "genes[acronym$eq'",
                        gene,
                        "']"])
    print(build_string)
    response = requests.get(build_string)
    rJson=response.json()
    imageDataSet_ids[gene]=rJson['msg'][0]['specimen_id']

imageDataSet_ids
```

```python
import io
from PIL import Image
import numpy as np

response = requests.get("http://api.brain-map.org/api/v2/image_download/74434404?downsample=4")
response.raise_for_status()
with io.BytesIO(response.content) as f:
    with Image.open(f) as img:
        myImage=np.array(img)
```

```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.imshow(myImage)
plt.show()
```
