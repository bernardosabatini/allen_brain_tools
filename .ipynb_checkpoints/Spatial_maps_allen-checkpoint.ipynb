# Initialize and find the ImageDataSet ids for the genes of interest

```python
import requests
import json

genes=['Adora2a', 'Tph2', 'Slc17a7']
section_orientation='coronal'

imageDataSet_ids={}

for gene in genes:
    urlString=''.join(['http://api.brain-map.org/api/v2/data/',
                      'query.json?criteria=model::SectionDataSet,',
                      'rma::criteria,',
                      "[failed$eq'false'],",
                      "products[abbreviation$eq'Mouse'],",
                      "plane_of_section[name$eq'",
                      section_orientation,
                      "'],",
                      "genes[acronym$eq'",
                        gene,
                        "']"])
    print(gene)
    response = requests.get(urlString)
    rJson=response.json()
    print(rJson)
    imageDataSet_ids[gene]=rJson['msg'][0]['id']

imageDataSet_ids
```

# Get the reference image and plot it

```python
import io
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

def getImage(image_id, downSample, display):
    urlString=''.join(["http://api.brain-map.org/api/v2/image_download/",
                                 str(image_id),
                                 "?downsample=",
                                 str(downSample)])

    response = requests.get(urlString)

    response.raise_for_status()
    with io.BytesIO(response.content) as f:
        with Image.open(f) as img:
            newImage=np.array(img)
    if (display==1):
        fig = plt.figure()
        ax = fig.add_subplot(1,1,1)
        ax.imshow(newImage)
        plt.show()        
        
    return newImage
```

```python
zeroImage_id=74434404
downSample=1;
zeroImage=getImage(zeroImage_id, downSample, 1)
```

### Get center pixel

```python
zeroImageSize=zeroImage.shape
xC=round(zeroImageSize[1]/2)
yC=round(zeroImageSize[0]/2)
xC_raw=xC*(downSample+1)
yC_raw=yC*(downSample+1)

```

# Find the x,y,z anatomoical coordinate in the reference atlas that matches with the choosen point (center) in the zeroImage
This is go from an (x,y) in a image to the (x,y,z) coordinate in a reference dataset
The plan will be to do this for each of N images that span the structure of interest, the (x,y) set to the center of the image

```python
urlString="".join(["http://api.brain-map.org/api/v2/image_to_reference/",
                                str(zeroImage_id),
                                ".json?x=",
                                str(xC_raw),
                                "&y=",
                                str(yC_raw)])
print(urlString)
response = requests.get(urlString)

print(response)
reference_equiv=response.json();
reference_equiv

```

## Get the x,y,z (now true anatomical coordinates) of the corresponding point in the reference atlas.

```python
zeroImageToReference_xyz=reference_equiv['msg']['image_to_reference']
print (zeroImageToReference_xyz)
print(zeroImageToReference_xyz['x'])
```

## Get the sections number and coordinates within each image set of each Gene that best matches with the x,y,z coordinate of the reference atlas
